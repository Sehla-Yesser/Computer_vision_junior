## 🔧 1. Setup & Configuration

In this section, we:
- Import all required libraries
- Set random seeds for reproducibility
- Define hyperparameters (image size, batch size, epochs, etc.)
- Configure the training environment

````python
# Import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import cv2

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============ Configuration Parameters ============
IMG_SIZE = 48          # Input image size (48x48 pixels)
BATCH_SIZE = 64        # Number of images per batch
EPOCHS = 50            # Maximum training epochs
LEARNING_RATE = 0.001  # Initial learning rate

# Emotion labels (7 classes)
EMOTIONS = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

print("=" * 70)
print("🎭 Facial Emotion Recognition using CNN")
print("=" * 70)
print(f"\n📊 Configuration:")
print(f"  • Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"  • Batch Size: {BATCH_SIZE}")
print(f"  • Max Epochs: {EPOCHS}")
print(f"  • Learning Rate: {LEARNING_RATE}")
print(f"  • Number of Classes: {len(EMOTIONS)}")
print("=" * 70)
````

## 📂 2. EmotionRecognitionCNN Class

This is the main class that encapsulates all functionality:
- Dataset exploration
- Data preprocessing and augmentation
- Model building
- Training
- Evaluation
- Visualization

Let's define the class with all its methods:

````python
class EmotionRecognitionCNN:
    """
    Main class for Facial Expression Recognition using CNN
    
    This class handles:
    - Dataset loading and exploration
    - Data augmentation
    - CNN model creation
    - Training with callbacks
    - Evaluation and visualization
    """
    
    def __init__(self):
        """Initialize the emotion recognition system"""
        self.model = None
        self.history = None
        self.train_generator = None
        self.val_generator = None
        self.test_generator = None
        print("✓ EmotionRecognitionCNN initialized")
````

### 2.1 Dataset Exploration Method

This method explores the dataset structure and counts images per emotion class:

````python
    def explore_dataset(self, base_path):
        """
        Explore the dataset structure and count images per class
        
        Args:
            base_path: Path to the dataset root folder
            
        Returns:
            list: List of emotion class names found
        """
        print("\n📂 Exploring Dataset Structure...")
        print("-" * 70)
        
        # Define paths
        train_path = os.path.join(base_path, 'train')
        test_path = os.path.join(base_path, 'test')
        
        print(f"Training path: {train_path}")
        print(f"Testing path: {test_path}")
        
        # List emotion folders in training
        if os.path.exists(train_path):
            emotions = sorted(os.listdir(train_path))
            print(f"\n📊 Detected {len(emotions)} emotion classes:")
            
            for emotion in emotions:
                emotion_path = os.path.join(train_path, emotion)
                if os.path.isdir(emotion_path):
                    num_images = len([f for f in os.listdir(emotion_path) 
                                    if f.endswith(('.png', '.jpg', '.jpeg'))])
                    print(f"  • {emotion.capitalize()}: {num_images} images")
            
            return emotions
        else:
            print("⚠️  Training folder not found!")
            return None
````

### 2.2 Data Generators with Augmentation

**Data Augmentation** helps prevent overfitting by creating variations of training images:
- Random rotations (±15°)
- Shifts (10% horizontal and vertical)
- Horizontal flips
- Zoom (±10%)
- Shear transformations

````python
    def setup_data_generators(self, base_path, validation_split=0.2):
        """
        Setup ImageDataGenerator for loading and augmenting images
        
        Args:
            base_path: Path to the dataset root folder
            validation_split: Fraction of training data to use for validation (default: 0.2)
            
        Returns:
            list: List of class names detected
        """
        print("\n🔄 Setting up Data Generators...")
        print("-" * 70)
        
        train_path = os.path.join(base_path, 'train')
        test_path = os.path.join(base_path, 'test')
        
        # Training data WITH augmentation
        train_datagen = ImageDataGenerator(
            rescale=1./255,                # Normalize pixel values to [0, 1]
            rotation_range=15,             # Random rotation ±15 degrees
            width_shift_range=0.1,         # Horizontal shift ±10%
            height_shift_range=0.1,        # Vertical shift ±10%
            horizontal_flip=True,          # Random horizontal flip
            zoom_range=0.1,                # Random zoom ±10%
            shear_range=0.1,               # Shear transformation
            fill_mode='nearest',           # Fill strategy for new pixels
            validation_split=validation_split
        )
        
        # Test data WITHOUT augmentation (only rescaling)
        test_datagen = ImageDataGenerator(rescale=1./255)
        
        # Create training generator
        self.train_generator = train_datagen.flow_from_directory(
            train_path,
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            color_mode='grayscale',        # Grayscale images (1 channel)
            class_mode='categorical',      # One-hot encoded labels
            subset='training',
            shuffle=True
        )
        
        # Create validation generator
        self.val_generator = train_datagen.flow_from_directory(
            train_path,
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            color_mode='grayscale',
            class_mode='categorical',
            subset='validation',
            shuffle=False
        )
        
        # Create test generator
        self.test_generator = test_datagen.flow_from_directory(
            test_path,
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            color_mode='grayscale',
            class_mode='categorical',
            shuffle=False
        )
        
        # Get class names
        class_names = list(self.train_generator.class_indices.keys())
        print(f"\n✓ Class names detected: {class_names}")
        print(f"✓ Training samples: {self.train_generator.samples}")
        print(f"✓ Validation samples: {self.val_generator.samples}")
        print(f"✓ Test samples: {self.test_generator.samples}")
        
        return class_names
````

### 2.3 CNN Model Architecture

Our deep CNN consists of **4 convolutional blocks** followed by **fully connected layers**:

**Architecture Overview:**
```
Input (48x48x1) 
    ↓
Conv Block 1 (64 filters) → MaxPool → Dropout
    ↓
Conv Block 2 (128 filters) → MaxPool → Dropout
    ↓
Conv Block 3 (256 filters) → MaxPool → Dropout
    ↓
Conv Block 4 (512 filters) → MaxPool → Dropout
    ↓
Flatten → Dense (512) → Dropout
    ↓
Dense (256) → Dropout
    ↓
Output (7 classes with Softmax)
```

**Key Features:**
- **Batch Normalization**: Stabilizes and accelerates training
- **ReLU Activation**: Non-linearity for feature learning
- **MaxPooling**: Reduces spatial dimensions
- **Dropout**: Prevents overfitting

````python
    def build_cnn_model(self, num_classes):
        """
        Build deep CNN architecture for emotion recognition
        
        Architecture: 4 Convolutional Blocks + 2 Dense Layers
        - Each conv block: 2x Conv2D → BatchNorm → ReLU → MaxPool → Dropout
        - Dense layers with batch normalization and dropout
        
        Args:
            num_classes: Number of emotion classes to predict
            
        Returns:
            Compiled Keras model
        """
        print("\n🏗️  Building CNN Model...")
        print("-" * 70)
        
        model = models.Sequential(name='EmotionCNN')
        
        # ========== Block 1: 64 filters ==========
        model.add(layers.Conv2D(64, (3, 3), padding='same', 
                                input_shape=(IMG_SIZE, IMG_SIZE, 1)))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Conv2D(64, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling2D(pool_size=(2, 2)))
        model.add(layers.Dropout(0.25))
        
        # ========== Block 2: 128 filters ==========
        model.add(layers.Conv2D(128, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Conv2D(128, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling2D(pool_size=(2, 2)))
        model.add(layers.Dropout(0.25))
        
        # ========== Block 3: 256 filters ==========
        model.add(layers.Conv2D(256, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Conv2D(256, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling2D(pool_size=(2, 2)))
        model.add(layers.Dropout(0.25))
        
        # ========== Block 4: 512 filters ==========
        model.add(layers.Conv2D(512, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Conv2D(512, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling2D(pool_size=(2, 2)))
        model.add(layers.Dropout(0.25))
        
        # ========== Fully Connected Layers ==========
        model.add(layers.Flatten())
        model.add(layers.Dense(512))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Dropout(0.5))
        
        model.add(layers.Dense(256))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.Dropout(0.5))
        
        # ========== Output Layer ==========
        model.add(layers.Dense(num_classes, activation='softmax'))
        
        self.model = model
        
        # Compile model
        model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print("✓ Model built successfully!")
        print(f"\nTotal Parameters: {model.count_params():,}")
        model.summary()
        
        return model
````

### 2.4 Training with Callbacks

**Training Callbacks** help optimize the training process:

1. **EarlyStopping**: Stops training if validation loss doesn't improve for 10 epochs
2. **ReduceLROnPlateau**: Reduces learning rate by 50% if validation loss plateaus
3. **ModelCheckpoint**: Saves the best model based on validation accuracy

````python
    def train_model(self):
        """
        Train the CNN model with callbacks for optimization
        
        Uses:
        - EarlyStopping to prevent overfitting
        - ReduceLROnPlateau for adaptive learning rate
        - ModelCheckpoint to save best weights
        
        Returns:
            Training history object
        """
        print("\n🚀 Starting Training...")
        print("-" * 70)
        
        # Create output directory
        os.makedirs('emotion_recognition', exist_ok=True)
        
        # ========== Callbacks ==========
        # Stop training if no improvement
        early_stop = EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True,
            verbose=1
        )
        
        # Reduce learning rate when stuck
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=1
        )
        
        # Save best model
        checkpoint = ModelCheckpoint(
            'emotion_recognition/best_model.h5',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=1
        )
        
        # Calculate steps per epoch
        steps_per_epoch = self.train_generator.samples // BATCH_SIZE
        validation_steps = self.val_generator.samples // BATCH_SIZE
        
        # Train model
        self.history = self.model.fit(
            self.train_generator,
            steps_per_epoch=steps_per_epoch,
            validation_data=self.val_generator,
            validation_steps=validation_steps,
            epochs=EPOCHS,
            callbacks=[early_stop, reduce_lr, checkpoint],
            verbose=1
        )
        
        print("\n✓ Training completed!")
        
        return self.history
````

### 2.5 Model Evaluation

Evaluate the trained model on the test set and generate detailed metrics:

````python
    def evaluate_model(self):
        """
        Evaluate model performance on test set
        
        Generates:
        - Test accuracy and loss
        - Classification report (precision, recall, F1-score)
        - Predictions for confusion matrix
        
        Returns:
            tuple: (predictions, true_labels, class_names)
        """
        print("\n📊 Evaluating Model on Test Set...")
        print("-" * 70)
        
        # Reset test generator
        self.test_generator.reset()
        
        # Evaluate
        test_loss, test_accuracy = self.model.evaluate(
            self.test_generator,
            steps=self.test_generator.samples // BATCH_SIZE,
            verbose=1
        )
        
        print(f"\n{'='*50}")
        print(f"📈 Test Results:")
        print(f"  • Test Loss: {test_loss:.4f}")
        print(f"  • Test Accuracy: {test_accuracy*100:.2f}%")
        print(f"{'='*50}")
        
        # Get predictions
        self.test_generator.reset()
        predictions = self.model.predict(
            self.test_generator,
            steps=self.test_generator.samples // BATCH_SIZE,
            verbose=1
        )
        
        y_pred = np.argmax(predictions, axis=1)
        y_true = self.test_generator.classes[:len(y_pred)]
        
        # Get class names
        class_names = list(self.test_generator.class_indices.keys())
        
        # Classification report
        print("\n📋 Classification Report:")
        print("="*70)
        print(classification_report(y_true, y_pred, target_names=class_names))
        
        return y_pred, y_true, class_names
````

### 2.6 Visualization Methods

Methods for visualizing training progress and results:

````python
    def plot_training_history(self):
        """Plot training and validation accuracy/loss curves"""
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        
        # Accuracy plot
        axes[0].plot(self.history.history['accuracy'], label='Train Accuracy', linewidth=2)
        axes[0].plot(self.history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
        axes[0].set_xlabel('Epoch', fontsize=12)
        axes[0].set_ylabel('Accuracy', fontsize=12)
        axes[0].set_title('Model Accuracy Over Time', fontsize=14, fontweight='bold')
        axes[0].legend(fontsize=10)
        axes[0].grid(True, alpha=0.3)
        
        # Loss plot
        axes[1].plot(self.history.history['loss'], label='Train Loss', linewidth=2)
        axes[1].plot(self.history.history['val_loss'], label='Val Loss', linewidth=2)
        axes[1].set_xlabel('Epoch', fontsize=12)
        axes[1].set_ylabel('Loss', fontsize=12)
        axes[1].set_title('Model Loss Over Time', fontsize=14, fontweight='bold')
        axes[1].legend(fontsize=10)
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('emotion_recognition/training_history.png', dpi=100, bbox_inches='tight')
        plt.show()
        
    def plot_confusion_matrix(self, y_true, y_pred, class_names):
        """
        Plot confusion matrix heatmap
        
        Args:
            y_true: True labels
            y_pred: Predicted labels
            class_names: List of class names
        """
        cm = confusion_matrix(y_true, y_pred)
        
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                    xticklabels=[name.capitalize() for name in class_names],
                    yticklabels=[name.capitalize() for name in class_names],
                    cbar_kws={'label': 'Count'})
        plt.xlabel('Predicted Emotion', fontsize=12, fontweight='bold')
        plt.ylabel('True Emotion', fontsize=12, fontweight='bold')
        plt.title('Confusion Matrix - Emotion Classification', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.savefig('emotion_recognition/confusion_matrix.png', dpi=100, bbox_inches='tight')
        plt.show()
    
    def visualize_predictions(self, num_samples=16):
        """
        Visualize sample predictions from test set
        
        Args:
            num_samples: Number of samples to display (default: 16)
        """
        # Reset generator
        self.test_generator.reset()
        
        # Get a batch
        images, labels = next(self.test_generator)
        
        # Predict
        predictions = self.model.predict(images[:num_samples], verbose=0)
        
        # Get class names
        class_names = list(self.test_generator.class_indices.keys())
        
        # Plot
        fig, axes = plt.subplots(4, 4, figsize=(12, 12))
        axes = axes.ravel()
        
        for i in range(min(num_samples, len(images))):
            # True and predicted labels
            true_label = class_names[np.argmax(labels[i])]
            pred_label = class_names[np.argmax(predictions[i])]
            confidence = np.max(predictions[i]) * 100
            
            # Plot image
            axes[i].imshow(images[i].squeeze(), cmap='gray')
            axes[i].axis('off')
            
            # Color: green if correct, red if wrong
            color = 'green' if true_label == pred_label else 'red'
            axes[i].set_title(
                f'True: {true_label.capitalize()}\nPred: {pred_label.capitalize()} ({confidence:.1f}%)',
                color=color, fontsize=9, fontweight='bold'
            )
        
        plt.tight_layout()
        plt.savefig('emotion_recognition/sample_predictions.png', dpi=100, bbox_inches='tight')
        plt.show()
    
    def save_model(self):
        """Save the trained model to disk"""
        self.model.save('emotion_recognition/emotion_recognition_model.h5')
        print("\n✓ Model saved to 'emotion_recognition/emotion_recognition_model.h5'")
````

## 🎬 3. Main Execution Flow

Now let's define the main execution function that ties everything together:

````python
def main():
    """
    Main execution function for training the emotion recognition model
    
    Steps:
    1. Explore dataset
    2. Setup data generators
    3. Build CNN model
    4. Train the model
    5. Evaluate on test set
    6. Visualize results
    7. Save the model
    """
    # Dataset path - UPDATE THIS to match your Kaggle path
    BASE_PATH = '/kaggle/input/fer2013'
    
    # Alternative paths (uncomment if needed):
    # BASE_PATH = '/kaggle/input/emotion-detection-fer'
    # BASE_PATH = 'path/to/your/dataset'
    
    print("\n" + "="*70)
    print("🎭 FACIAL EXPRESSION RECOGNITION - CNN TRAINING")
    print("="*70)
    
    # Initialize the CNN system
    emotion_cnn = EmotionRecognitionCNN()
    
    # Step 1: Explore dataset
    class_names = emotion_cnn.explore_dataset(BASE_PATH)
    
    if class_names is None:
        print("\n⚠️  ERROR: Dataset not found!")
        print("\n📁 Expected structure:")
        print("  /kaggle/input/YOUR-DATASET/")
        print("    ├── train/")
        print("    │   ├── angry/")
        print("    │   ├── happy/")
        print("    │   └── ... (other emotions)")
        print("    └── test/")
        print("        ├── angry/")
        print("        ├── happy/")
        print("        └── ... (other emotions)")
        return
    
    # Step 2: Setup data generators
    class_names = emotion_cnn.setup_data_generators(BASE_PATH)
    
    # Step 3: Build CNN model
    emotion_cnn.build_cnn_model(num_classes=len(class_names))
    
    # Step 4: Train the model
    history = emotion_cnn.train_model()
    
    # Step 5: Evaluate on test set
    y_pred, y_true, class_names = emotion_cnn.evaluate_model()
    
    # Step 6: Visualize results
    print("\n📊 Generating visualizations...")
    emotion_cnn.plot_training_history()
    emotion_cnn.plot_confusion_matrix(y_true, y_pred, class_names)
    emotion_cnn.visualize_predictions(num_samples=16)
    
    # Step 7: Save the model
    emotion_cnn.save_model()
    
    print("\n" + "="*70)
    print("✅ TRAINING COMPLETE!")
    print("="*70)
    print("\n📁 Output files saved in 'emotion_recognition/' directory:")
    print("  • best_model.h5")
    print("  • emotion_recognition_model.h5")
    print("  • training_history.png")
    print("  • confusion_matrix.png")
    print("  • sample_predictions.png")
    print("\n" + "="*70)
````

## 🔧 6. Troubleshooting

### Common Issues:

**1. Dataset Not Found**
- Verify the `BASE_PATH` variable points to your dataset
- Check folder structure matches expected format

**2. Out of Memory**
- Reduce `BATCH_SIZE` (try 32 or 16)
- Reduce model complexity (fewer filters)

**3. Overfitting (High train accuracy, low val accuracy)**
- Already using dropout and batch normalization
- Could add more data augmentation
- Try early stopping with smaller patience

**4. Slow Training**
- Enable GPU in Kaggle (Settings → Accelerator → GPU)
- Reduce image size if needed
- Use fewer epochs initially for testing
